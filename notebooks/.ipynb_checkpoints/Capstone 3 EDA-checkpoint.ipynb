## Plant Efficiency

```python
# We can specify what dataset we are giving the below function by specifying data_type as either:
# 'gd' (i.e. generation data) or 'wsd' (i.e weather sensor data)

def time_generation(data):
    grouped_df = data.groupby(['time'])
    df = grouped_df.agg({'dc_power':['mean','std'],
                             'ac_power':['mean','std'],
                             'daily_yield':['mean','std'],
                             'total_yield':['median','std']}).reset_index()
    df.columns = ['time','avg_dc_power','se_dc_power','avg_ac_power','se_ac_power',
                     'avg_daily_yield','se_daily_yield','median_total_yield','se_total_yield']
    return df

def time_weather(data):
    grouped_df = data.groupby(['time'])
    df = grouped_df.agg({'ambient_temperature': ['mean', 'std'],
                             'module_temperature': ['mean', 'std'],
                             'irradiation': ['mean', 'std']}).reset_index()
    df.columns = ['time', 'avg_ambient_temperature','se_ambient_temperature','avg_module_temperature',
                 'se_module_temperature','avg_irradiation','se_irradiation']
    return df
```

```python
grouped1 = plant_df1.groupby(['date','time']).agg('sum').reset_index()
grouped2 = plant_df2.groupby(['date','time']).agg('sum').reset_index()
```

```python
# PLANT 1 GENERATION DATA GROUPED BY TIME
gd_t1 = time_generation(grouped1)
# PLANT 2 GENERATION DATA GROUPED BY TIME
gd_t2 = time_generation(grouped2)
# gd_time_data = [gd_t1, gd_t2]
```

```python
# As above we pass plants as a list: [plant1-data,plant2-data]
def avg_plots(plant1,plant2,var,style='g-'):
    x1 = plant1['time'].astype(str)
    y1 = plant1['avg_'+var]  
    se1 = plant1['se_'+var]

    x2 = plant2['time'].astype(str)
    y2 = plant2['avg_'+var]
    se2 = plant2['se_'+var]
    
    fig,ax = plt.subplots(1,2,sharey=True)
    fig.suptitle(var+' averages')

    ax[0].plot(x1,y1,style,linewidth=2)
    ax[1].plot(x2,y2,style,linewidth=2)

    ax[0].set_title('Plant 1')
    ax[1].set_title('Plant 2')
    
    ax[0].fill_between(x1,y1-se1,y1+se1)
    ax[1].fill_between(x2,y2-se2,y2+se2)

    ax[0].set_xlabel('time')
    ax[1].set_xlabel('time')
    
    ax[0].set_xticks(np.arange(0,100,10))
    ax[1].set_xticks(np.arange(0,100,10))
    
    ax[0].tick_params(axis='x',rotation=90)
    ax[1].tick_params(axis='x',rotation=90)

    ax[0].set_ylabel('avg '+var)

    plt.tight_layout();
```

```python
avg_plots(gd_t1,gd_t2,'ac_power')
avg_plots(gd_t1,gd_t2,'dc_power')
avg_plots(gd_t1,gd_t2,'daily_yield')
```
